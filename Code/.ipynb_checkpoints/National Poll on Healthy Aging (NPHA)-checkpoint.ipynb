```python
def EDA(df):
    # Drop rows with missing values
    df_cleaned = df.dropna()

    # Descriptive statistics
    descriptive_stats = df_cleaned.describe()

    # Inferential statistics (example: hypothesis test for Age)
    age = df_cleaned['Age']
    age_mean = age.mean()
    t_stat, p_value = stats.ttest_1samp(age, popmean=age_mean)

    # Covariance
    covariance_matrix = df_cleaned.cov()

    # Correlation
    correlation_matrix = df_cleaned.corr()

    # Correlation Heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
    
    # Display pairplot for numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    sns.pairplot(df[numerical_cols])
    plt.title('Pairplot of Numerical Columns')
    plt.show()
    
    # Histograms and Boxplots for numerical variables
    numeric_columns = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
    for column in numeric_columns:
        column_mean = df_cleaned[column].mean()
        column_mode = df_cleaned[column].mode().values[0] 
        column_median = df_cleaned[column].median()
        column_q1 = df_cleaned[column].quantile(0.25)
        column_q3 = df_cleaned[column].quantile(0.75)

        print(f"--- {column} ---")
        print(f"Mean: {column_mean}")
        print(f"Mode: {column_mode}")
        print(f"Median: {column_median}")
        print(f"Q1: {column_q1}")
        print(f"Q3: {column_q3}")

        # Histogram
        plt.figure(figsize=(8, 6))
        sns.histplot(df_cleaned[column], kde=True)
        plt.title(f'Histogram of {column}')
        plt.show()

        # Boxplot
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df_cleaned[column])
        plt.title(f'Boxplot of {column}')
        plt.show()

    return descriptive_stats, t_stat, p_value, covariance_matrix, correlation_matrix
```

```python
def manual_splitting(df, target_column, test_size=0.2, validation_size=0.2, random_state=32):
    X = df.drop(columns=[target_column])
    y = df[target_column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=validation_size, random_state=random_state)

    return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
#------------------------------Splitting Data using Cross Validation-----------------------------
def cross_validation(df, target_column, n_splits=2, random_state=42):
    X = df.drop(columns=[target_column])
    y = df[target_column]
    skf = StratifiedKFold(n_splits=n_splits, random_state=random_state, shuffle=True)

    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        yield X_train, X_test, y_train, y_test
```

```python
#----------In this model, we predict the need for prescription sleep medication based on various features---
def model_selection(df, target_column, k_values=[3, 5, 7, 9]):
    X_train, X_val, X_test, y_train, y_val, y_test = manual_splitting(df, target_column)
    
    # Feature Engineering (if needed)
    # Hyperparameter Tuning
    best_k = None
    best_accuracy = 0
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_val)
        accuracy = accuracy_score(y_val, y_pred)
        if accuracy > best_accuracy:
            best_k = k
            best_accuracy = accuracy
    
    # Train the final model with the best k
    knn_final = KNeighborsClassifier(n_neighbors=best_k)
    knn_final.fit(X_train, y_train)
    
    # Get the final accuracy on the test set
    y_pred_test = knn_final.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred_test)

    return best_k, best_accuracy, test_accuracy

best_k, best_accuracy, test_accuracy = model_selection(Data_Transformation(Data_Load()), 'Prescription Sleep Medication')
print("Best K:", best_k)
print("Validation Accuracy with Best K:", best_accuracy)
print("Test Accuracy with Best K:", test_accuracy)
```

```python
# clalculationg classification metrics like accurary,precision,recall,roc
def classification_metrics(df, target_column, k=3):
    label_encoder = LabelEncoder()
    df_encoded = df.copy()
    df_encoded[target_column] = label_encoder.fit_transform(df[target_column])

    X_train, X_val, X_test, y_train, y_val, y_test = manual_splitting(df_encoded, target_column)
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_manual = knn.predict(X_test)

    accuracy_manual = accuracy_score(y_test, y_pred_manual)
    report_manual = classification_report(y_test, y_pred_manual,zero_division=1)
    roc_auc_manual = roc_auc_score(y_test, knn.predict_proba(X_test), multi_class='ovr')

    # Cross-validation
    X = df_encoded.drop(columns=[target_column])
    y = df_encoded[target_column]
    knn_cv = KNeighborsClassifier(n_neighbors=k)
    y_pred_cv = cross_val_predict(knn_cv, X, y, cv=5)
    
    accuracy_cv = accuracy_score(y, y_pred_cv)
    report_cv = classification_report(y, y_pred_cv, zero_division=1)
    roc_auc_cv = roc_auc_score(y, knn.predict_proba(X), multi_class='ovr')

    # ROC curve
    y_scores_manual = knn.predict_proba(X_test)
    fpr, tpr, _ = roc_curve(pd.get_dummies(y_test).values.ravel(), y_scores_manual.ravel())
    roc_auc_curve = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_curve)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc='lower right')
    plt.show()

    return accuracy_manual, report_manual, roc_auc_manual, accuracy_cv, report_cv, roc_auc_cv

# Call the classification metrics function
accuracy_manual, report_manual, roc_auc_manual, accuracy_cv, report_cv, roc_auc_cv = classification_metrics(Data_Transformation(Data_Load()), 'Prescription Sleep Medication')
print("Manual Splitting Accuracy:", accuracy_manual)
print("Manual Splitting Classification Report:\n", report_manual)
print("Manual Splitting ROC AUC:", roc_auc_manual)
print("\nCross-Validation Accuracy:", accuracy_cv)
print("Cross-Validation Classification Report:\n", report_cv)
print("Cross-Validation ROC AUC:", roc_auc_cv)
```

```python
def Master():
    # Load the dataset
    df = Data_Load()

    # Perform data transformation
    df_transformed = Data_Transformation(df)

    # Perform exploratory data analysis
    EDA(df_transformed)

    # Detect outliers and anomalies
    Detecting_outliers_Anomalies(df_transformed)

    # Split the data using manual splitting
    X_train_manual, X_val_manual, X_test_manual, y_train_manual, y_val_manual, y_test_manual = manual_splitting(Data_Transformation(df), 'Prescription Sleep Medication')

    # Train the model and get predictions for manual splitting
    knn = KNeighborsClassifier(n_neighbors=3)
    knn.fit(X_train_manual, y_train_manual)
    y_pred_manual = knn.predict(X_test_manual)

    # Get classification metrics for manual splitting
    accuracy_manual, report_manual, roc_auc_manual, accuracy_cv, report_cv, roc_auc_cv = classification_metrics(df_transformed, 'Prescription Sleep Medication')

    # Get cross-validation accuracy
    cv_accuracy = cross_validation(df_transformed, 'Prescription Sleep Medication')

    # Print the results for manual splitting
    print("Manual Splitting Test Accuracy:", accuracy_manual)
    print("Classification Report (Manual Splitting):")
    print(report_manual)
    print("ROC AUC (Manual Splitting):", roc_auc_manual)
    print()
    # Print the cross-validation accuracy
    print("Cross-Validation Mean Accuracy:", cv_accuracy)
```

```python
# Master function that's controlling pipeline of fucntions
Master()
```
