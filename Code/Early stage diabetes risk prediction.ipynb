```python
# this dataset don;t have any missing value as i have checked it
def Data_Cleaning(df):
    return df.isnull().sum() 
```

```python
#data transformation by encoding categorical variables and scaling numerical variables
#  encoding the specified categorical columns ('Gender', 'Polyuria', 'Polydipsia', etc.) into numerical values.
# scales the 'Age' column to have a mean of 0 and a standard deviation of 1.
def Data_Transformation(df):
    # Encode categorical variables
    label_encoder = LabelEncoder()
    categorical_columns = ['Gender', 'Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness',
                           'Polyphagia', 'Genital thrush', 'visual blurring', 'Itching', 'Irritability',
                           'delayed healing', 'partial paresis', 'muscle stiffness', 'Alopecia', 'Obesity']
    for column in categorical_columns:
        df[column] = label_encoder.fit_transform(df[column])

    # Scale numerical variables
    scaler = StandardScaler()
    numerical_columns = ['Age']
    df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

    return df
```

```python
def EDA(df):
    # Summary statistics
    summary = df.describe()

    # Distribution of the target variable 'class'
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='class')
    plt.title('Distribution of Class')
    plt.show()

    # Pairplot for numerical variables
    sns.pairplot(df, vars=['Age'], hue='class', diag_kind='hist')
    plt.title('Pairplot of Age by Class')
    plt.show()

    # Correlation matrix
    correlation_matrix = df.corr()
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()
    
    # Display pairplot for numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    sns.pairplot(df[numerical_cols])
    plt.title('Pairplot of Numerical Columns')
    plt.show()
    
    # Boxplot for numerical variables by class
    numerical_columns = ['Age']
    for column in numerical_columns:
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df, x='class', y=column)
        plt.title(f'Boxplot of {column} by Class')
        plt.show()

    # Countplot for categorical variables
    categorical_columns = ['Gender', 'Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness',
                           'Polyphagia', 'Genital thrush', 'visual blurring', 'Itching', 'Irritability',
                           'delayed healing', 'partial paresis', 'muscle stiffness', 'Alopecia', 'Obesity']
    for column in categorical_columns:
        plt.figure(figsize=(8, 6))
        sns.countplot(data=df, x=column, hue='class')
        plt.title(f'Distribution of {column} by Class')
        plt.show()

    return summary
```

```python
#------------------------------Splitting Data using Manual way-----------------------------
def manual_splitting(df, target_column, test_size=0.2, validation_size=0.2, random_state=42):
    X = df.drop(columns=[target_column])
    y = df[target_column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=validation_size, random_state=random_state)

    return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
#------------------------------Splitting Data using Cross Validation-----------------------------
def cross_validation(df, target_column, k=3, cv=5, random_state=32):
    X = df.drop(columns=[target_column])
    y = df[target_column]
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=cv, scoring='accuracy')

    return scores.mean()
```

```python
# model selection fucntion, currently select KNN
def model_selection(df, target_column, k=3):
    X_train, X_val, X_test, y_train, y_val, y_test = manual_splitting(df, target_column)
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_val)
    test_accuracy_manual = accuracy_score(y_val, y_pred)

    cv_accuracy = cross_validation(df, target_column)

    return test_accuracy_manual, cv_accuracy
```

```python
# clalculationg classification metrics like accurary,precision,recall,roc
def classification_metrics(df, target_column, k=3):
    label_encoder = LabelEncoder()
    df_encoded = df.copy()
    df_encoded[target_column] = label_encoder.fit_transform(df[target_column])

    X_train, X_val, X_test, y_train, y_val, y_test = manual_splitting(df_encoded, target_column)
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_manual = knn.predict(X_test)

    accuracy_manual = accuracy_score(y_test, y_pred_manual)
    report_manual = classification_report(y_test, y_pred_manual)
    roc_auc_manual = roc_auc_score(y_test, y_pred_manual, multi_class='ovr')

    # Cross-validation
    X = df_encoded.drop(columns=[target_column])
    y = df_encoded[target_column]
    knn_cv = KNeighborsClassifier(n_neighbors=k)
    y_pred_cv = cross_val_predict(knn_cv, X, y, cv=5)
    
    accuracy_cv = accuracy_score(y, y_pred_cv)
    report_cv = classification_report(y, y_pred_cv)
    roc_auc_cv = roc_auc_score(y, y_pred_cv, multi_class='ovr')

    # ROC curve
    fpr, tpr, _ = roc_curve(y_test.ravel(), y_pred_manual.ravel())
    roc_auc_curve = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_curve)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc='lower right')
    plt.show()

    return accuracy_manual, report_manual, roc_auc_manual, accuracy_cv, report_cv, roc_auc_cv
```
